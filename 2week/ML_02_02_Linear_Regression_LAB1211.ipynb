## 1.Linear Regression의 이해



````
Linear Regression은 숫자로된 데이터로 예측하는 것으로 target이 연속형 이어야 함
````

### 1-01. 단순 선형 회귀

````
한 포장마차 주인이 한 해 동안 매일 기온과 오뎅 판매 수량을 계속 기록했다.
그랬더니"기온과 오뎅 판매 수량"사이에 선형 관계가 있었다.
이 주인은 위의 데이터로 만들어진 일차 함수식을 장사에 이용해보기로 했다.
이 주인은 일기예보를 통해 알게 된 내일의 기온을 일차 함수식에 대입했다.
그랬더니 내일 대략 몇 개가 팔릴 것인지 예측할 수 있었다.
그래서 이 주인은 이 수량에 맞춰서 재료를 구입했고,
덕분에 신선하고도 가장 맛있는 오뎅을 판매할 수 있었다.
그리고 판매 예상량을 적절히 예측했기 때문에 아깝게 버리는 오뎅들도 거의 없었다.
소문을 들은 여러 사람들이 이 포장마차에 찾아오게 되었고 결국 엄청난 수익을 내게 되었다.
````

- 단순 선형 회귀 : 하나의 특성(Feature)를 가지고 Target을 예측하기 위한 선형 함수를 찾는 것 
- 선형 함수 : X, y 의 관계를 설명하는 함수
  - X : sample의 특성 값
  - y : label/target 값
- 선형 함수를 사용하여 새로운 sample의 특성 값이 주어졌을 때 그에 해당하는 label/target 값을 "예측"해 낼 수 있음
- 단순 선형 회귀를 위한 일반화된 예측 함수

### 1-02. 다중 선형 회귀(Multivariate Linear Regression)


````
회귀모델을 만들 때, 단 하나의 특성만 가지고 있다면 충분한 예측능력을 보이는 모델을 만들기는 상당히 어렵다. 
위에서 오뎅 판매 수량을 단 하나의 특성, 그날의 기온으로 예측했지만
사실상 더 많은 특성들을 고려해야만 제대로 된 결과를 예측해 낼 수 있을 것이다. 
요일마다 판매량이 다를 수도 있고, 기온 뿐만 아니라 비가 오거나 눈이 오는 상황도 판매량에 영향을 미칠 것이다.
또한 티비 프로그램에 오뎅을 맛있게 먹는 장면이 나왔다면 판매량이 많아질 수 있다. 
이러한 특성을 동시에 고려해야만 좀 더 정확한 예측 모델을 만들 수 있기 때문에, 
대부분의 경우에 하나의 특성이 아닌 여러 개의 특성을 활용해서 회귀모델을 만든다. 
이것이 바로 "다중 선형 회귀"다. 
````

- 다중 선형 회귀 : 여러 개의 특성을 활용해서 Target을 예측하는 회귀 모델을 만듦
- 다중 선형 회귀를 위한 일반화된 예측 함수
> image.png
- p : sample의 개수
- 다중 선형 회귀에서는 특성이 p+1 개이고, 가중치도 p+1 개이다
- p+1 개의 특성과 label/target의 관계를 잘 설명해 낼 수 있는 가중치들과 b를 찾아야 한다
- 특성이 두 개면 평면, 더 높은 차원에서는 초평면(hyperplane)이 된다

### 1-03. 가중치, 편향을 찾는 방법
- 실제값(label/target 값)과 예측값 사이의 평균 제곱 오차를 최소화하는 w와 b를 찾는다
- MSE(Mean Squared Error) : 실제값과 예측값 사이의 평균 제곱 오차
- 실제값과 예측값의 차이가 작을 수록 예측 성능이 좋기 때문에 많이 사용함
> image.png
- **MSE를 최소화하는 파라미터 w 와 b 를 찾는 것이 선형 회귀의 목적**이다. 

### 1-04. Feature의 수와 과대적합(Overfitting)
- 주어진 샘플들의 특성값과 라벨값의 관계를 필요이상 너무 자세하고 복잡하게 분석
- 샘플에 심취!
- 새로운 데이터가 주어졌을 때 제대로 예측해내기 어려울 수 있음
- 해결 방법으로 릿지(Ridge)와 라쏘(Lasso) 방법이 있음

### 1-05. L1, L2, L∞ norm
- norm : 선형대수학에서 벡터의 크기(magnitude) 또는 길이(length)를 측정하는 방법
- L1 norm (= Manhattan norm)
   - 벡터의 모든 성분의 절대값을 더함
- L2 norm (= Euclidean norm)
   - 출발점에서 도착점까지의 거리를 직선거리로 측정함
- L∞ norm (= max norm)
   - 벡터 성분들의 절대값 중에서 가장 큰 값으로 측정

### 1-06. 라쏘(Lasso) - L1 norm
- w : 가중치, b : 편향
- 기존의 선형 회귀에서 적절한 w, b를 찾아내는 것이 목적
- 라쏘는 추가 제약 조건을 덧붙임
  - MSE가 최소가 되게하는 w, b를 찾는 동시에 **w의 절대값들의 합이 최소**가 되게 해야함
  - **w의 모든 원소가 0이 되거나 0에 가깝게 되게 해야 함 => 불필요 특성 제거**
  - 어떤 특성은 모델을 만들 때 사용되지 않게 됨
  - **어떤 벡터의 요소들의 절대값들의 합은 L1-norm임**
- 즉, 라쏘는 간단히 말해 L1-norm 패널티를 가진 선형 회귀 방법!
> image.png 

- m : 가중치의 개수
- α : penalty의 효과를 조절해주는 파라미터
   - 값이 커지면 패널티 영향력이 커져서 0이되는 w가 많이 발생한다 (특성 반영이 제대로 되지 않을 수 있음) 
   - 값이 작아져서 0이되면 penalty 없는 선형 회귀와 같다
   - 작은 것부터 조금씩 크게 적용해 보아서 적절한 값을 찾는다 
- MSE와 penalty항의 합이 최소가 되게 하는 w, b를 찾는 것이 라쏘의 목적이다!
> image.png

- 라쏘(Lasso)의 장점
  - 제약 조건을 통해 일반화된 모형을 찾는다
  - 가중치들이 0이 되게 하므로써 그에 해당하는 특성들을 제외해준다.
  - 모델에서 가장 중요한 특성이 무엇인지 알게되는 등 모델 해석력이 좋아진다.

### 1-07. 릿지(Ridge) - L2-norm
- 패널티 항으로 **L2-norm 패널티**를 가진다. 
- 라쏘는 가중치들이 0이 되지만, 릿지의 가중치들은 0에 가까워질 뿐 0이 되지는 않는다.
- 특성이 많은데 그 중 **일부분만 중요하다면 라쏘**가 특성의 **중요도가 전체적으로 비슷하다면 릿지**가 좀 더 괜찮은 모델을 찾아줄 것이다.

### 2-01. 최소제곱법(Least Square)
- Residual(잔차) : 실제 값과 가설에 의한 예측 값의 차이
- RSS(Residual Sum of Square) : 오차(잔차)의 제곱의 합  
- 최소제곱법은 RSS가 가장 작게 되는 계수를 구함
- 이상치(Outlier)에 취약함
- Feature와 Sample의 수에 비례하여 계산 비용이 높아짐

image.png

````
# Error 와 Residual 의 차이
- 모집단에서 회귀식을 얻었다면, 그 회귀식을 통해 얻은 예측값과 실제 관측값의 차이가 "오차(Error)"이다
  - SSE(Error Sum of Square)라고 함, MSE(Mean Squared Error) : SSE를 표준화한 개념
- 표본집단에서 회귀식을 얻었다면, 그 회귀식을 통해 얻은 예측값과 실제 관측값의 차이가 "잔차(Residual)"이다
````

````
import numpy as np
import matplotlib.pyplot as plt

X = np.array([10,9,6,3,2]) 
y = np.array([90,80,60,50,30])

plt.plot(X, y, 'o')

a = np.sum( (X-X.mean()) * (y-y.mean())) /  np.sum( (X-X.mean())**2 )
b = y.mean() - a*X.mean()
print(f"a:{a}, b:{b}")

plt.plot(X, X*a+b, 'r-' )
plt.grid(True)
````

### 2-02. 최소제곱법 in sklearn
- ```sklearn.linear_model.LinearRegression(fit_intercept, normalize)```
    - OLS(Ordinary Least Sqaures) 최소 제곱 추정 방식 구현
    - `fit_intercept=True` : bias 사용 여부

    - `intercept_` : bias 절편

````
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

X = np.array([10,9,6,3,2]) 
y = np.array([90,80,60,50,30])
lr = LinearRegression()
lr.fit(X.reshape(-1,1), y)
print(lr.coef_, lr.intercept_)
plt.plot(X, y, 'o')
plt.plot(X, X*lr.coef_[0]+lr.intercept_, 'r-' )
````

### 2-03. 최소제곱법 Do it youself

````
## 초기 데이터
import numpy as np

x = np.arange(0, 3).reshape(3, -1)  # 2D
y = np.arange(1, 6, 2)  # 1D
print(x, y)  # Feature
````

````
## 최소제곱법 모델 적용
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(x, y)  

print(lm)
````

````
# 기울기 출력, gradient, coefficient
lm.coef_
````

````
# 절편 출력, intercept
lm.intercept_
````

````
# 위의 결과를 사용해 봅시다
def f(x) :
    return lm.coef_[0] * x + lm.intercept_, f'y = {lm.coef_[0]:.2f}x + {lm.intercept_:.2f}'

x = [[4]]  # 2D 이어야 함
y, equation = f(4)
print(equation)
print(f'x = {x[0]}, y = {y:.2f}')
````

- 예측한 값이 실제값과 유사한지 평가하는 척도
- 작을 수록 좋지만, 너무 작으면 과적합
- 이상치에 민감함
- ```sklearn.metrics.mean_squared_error(y_true, y_pred)```

### 3-05. 회귀 모델 평가 in sklearn

````
## 최소제곱법으로 구한 것
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

X = np.array([10,9,6,3,2]) 
y = np.array([90,80,60,50,30])
lr = LinearRegression()
lr.fit(X.reshape(-1,1), y)
print(lr.coef_, lr.intercept_)
plt.plot(X, y, 'o')
plt.plot(X, X*lr.coef_[0]+lr.intercept_, 'r-' )
plt.grid(True)
````

````
# test를 위한 데이터 필요함
X_test = np.array([4, 7, 8 ])
y_test = np.array([48,75, 78])
````

````
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

pred = lr.predict(X_test.reshape(-1,1))
r2 = r2_score(y_test, pred)
mae = mean_absolute_error(y_test, pred)
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
print(f'r2_score = {r2:.3f}, mae = {mae:.3f}')
print(f'mse = {mse:.3f}, rmse = {rmse:.3f}')
````

````
# 정확도(accuracy)
print(f'train score:{lr.score(X.reshape(-1,1), y):.3f}')
print(f'test score :{lr.score(X_test.reshape(-1,1), y_test):.3f}')
````

### 4-02. 선형 회귀 적용

````
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor

def regression_lab(data, model):
    # 스케일링 - 정규화 
    X_data = StandardScaler().fit_transform(data.data)
    X_train, X_test, y_train, y_test = train_test_split(X_data, data.target, test_size=0.3, random_state=1)
    lr = model()
    lr.fit(X_train, y_train)
    trainscore = lr.score(X_train, y_train)
    testscore  = lr.score(X_test, y_test)
    return lr, trainscore, testscore, X_test, y_test

lr_models = []
for model in (LinearRegression, SGDRegressor):
    lr, trainscore, testscore, X_test, y_test = regression_lab(boston, model)
    lr_models.append([lr, X_test, y_test])
    print(model.__name__)
    print(f'train score:{trainscore:.3f}   test score :{testscore:.3f}')
    print('-'* 20)
````

### 4-03. 회귀 모델 평가

````
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def print_lr_errors(model, X_test, y_test):
    pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, pred)
    mse = mean_squared_error(y_test, pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, pred)

    print(type(model).__name__)
    print(f'MAE:{mae:.3f}, MSE:{mse:.3f}, RMSE:{rmse:.3f}, R2:{r2:.3f}')
    print('coef:', [ f'{x:.3f}' for x in model.coef_])
    print('bias:', model.intercept_)
    print('-' * 50)

for model in lr_models :
    print_lr_errors(*model)
````

## 5.다항회귀(Polynomial Regression)
- 입력 변수와 출력 변수 간의 관계가 선형이 아닌 경우
- 1차식으로 표현 불가 -> 2차식으로 변형
- 데이터가 단순한 직선의 형태가 아닌 비선형 형태인 경우 선형 모델을 사용하여 비선형 데이터를 학습하기 위한 방법
  1. 각 특성의 거듭제곱을 새로운 특성으로 추가하고
  2. 확장된 특성을 포함한 데이터셋에 선형 모델을 훈련시키는 것
- ```from sklearn.preprocessing import PolynomialFeatures(degree, include_bias)```
  - degree : 차수
  - include_bias : 1로 채운 컬럼 추가 여부 
  - https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html

## 6.Feature수와 과적합(Overfitting)
- Feature == Degree == Dimension
- **차수(degree)가 높아 질수록 과적합(Overfitting) 발생가능성 높음**
- 과적합 : '학습 데이터'에 너무 적합되어 새로운 데이터에 대한 예측 성능 저하가 발생한 것
- 회귀 계수 크기 제어를 통해 해결함
- 규제(penalty)
   - 비용 함수에 규제 항목 추가
   - Min(RSS(W) + α*||W||)
   - if α==0 : 규제 없음, if α == ∞ : W = 0으로 최소화

- 다중공선성(Multicollinearity) : 독립변수들이 강한 선형관계에 있는 것

  image.png

- 다중공선성이 있으면, 잘못된 변수 해석, 예측 정확도 하락 등을 발생시킴


### 6-01. Polynomial Regression 과적합 예

````
# sin 함수에 차수를 1, 4, 18 차로 변화 하면서 관찰한다

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

def sin(X):
    return np.sin(1.5 * np.pi * X)
    #return np.sin( X)

m = 30
X = np.sort(np.random.rand(m))
y = sin(X) + np.random.randn(m) * 0.1

degrees = (1, 4, 18)
plt.figure(figsize=(15,5))

for i, dgr  in enumerate(degrees, start=1):

    polynomial = PolynomialFeatures(degree=dgr)
    lr = LinearRegression()    
    # 2가지 차례로 적용
    pipe = Pipeline([('polynomial', polynomial), ('linear_regression', lr)])
    pipe.fit(X.reshape(-1,1), y)
    
    X_test = np.linspace(0,1, 100)
    pred = pipe.predict(X_test[:, np.newaxis])
    plt.subplot(1, len(degrees), i)
    plt.plot(X_test, sin(X_test), 'g-', label='True Y')
    plt.plot(X_test, pred, 'r--', label='Predict')   # X_test[:, np.newaxis] => reshape(-1, 1) 과 동일
    plt.plot(X, y, 'b.', label='Sample')
    plt.xlim((0,1)); plt.ylim((-2,2))
    plt.legend()
    plt.title(f"Degree:{dgr}")
    
plt.show()
````

````
# PolynomialFeatures(degree=dgr)의 경우, include_bias=True로 동작하기 때문
# Degree 1 일때 Feature (X, 2)
# Degree 4 일때 Feature (X, 5) 
# Degree 5 일때 Feature (X, 19) 
````

````
- L1 norm (= Manhattan norm)
   - 벡터의 모든 성분의 절대값을 더함
- L2 norm (= Euclidean norm)
````

### 6-02. 랏소(Lasso) 모델 - Polynomial
- Least Absolute Shrinkage and Selection Operator
- L1 규제 계수 사용
- 가중치(W) 절대값의 합을 최소화
- 불필요한 회귀 계수를 급격히 감소, 0으로 만들어 제거함
- 특성(Feature) 선택에 유리함
- sklearn.linear_model.Lasso(alpha)
   - alpha = 1.0 : L1 규제 사용 
- https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Lasso.html

````
# alphas = (0, 0.005, 0.1, 1)로 하여 모델 적용

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso
import warnings   # Suppress Deprecation and Incorrect Usage Warnings
warnings.filterwarnings('ignore')

def sin(X):
    return np.sin(1.5 * np.pi * X)

dgr = 18
m = 30
X = np.sort(np.random.rand(m))
y = sin(X) + np.random.randn(m) * 0.1

alphas = (0.00001, 0.005, 0.1, 1)
plt.figure(figsize=(15,5))

for i, alpha  in enumerate(alphas):
    polynomial = PolynomialFeatures(degree=dgr)
    lasso = Lasso(alpha=alpha)
    pipe = Pipeline([('polynomial', polynomial), ('lasso', lasso)])
    pipe.fit(X.reshape(-1,1), y)
    
    X_test = np.linspace(0,1, 100)
    plt.subplot(1, len(alphas), i+1)
    plt.plot(X_test, sin(X_test), 'g-', label='True Y')
    plt.plot(X_test, pipe.predict(X_test[:, np.newaxis]), 'r--', label='Predict')
    plt.plot(X, y, 'b.', label='Sample')
    plt.xlim((0,1)); plt.ylim((-2,2))
    plt.legend()
    plt.title(f"Degree:{dgr}, Alpha:{alpha}")
    
plt.show()
````

### 6-03. 랏소(Lasso) 모델 - Multivariate/보스턴 주택
- 불필요한 회귀 계수 제거 실습
- 결과에서 0으로 표시되는 것이 제거 가능 한 것

````
from sklearn.datasets import load_boston
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score

boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)

X_data = boston.data
y_target = boston.target
X_train, X_test, y_train, y_test = train_test_split(X_data, y_target, test_size=0.3, random_state=1)

alphas = (0.0001, 0.1, 0.5, 1, 2)
coef_df = pd.DataFrame()

for i, alpha in enumerate(alphas):
    lasso = Lasso(alpha = alpha)
    lasso.fit(X_train, y_train)
    pred = lasso.predict(X_test)

    mse = mean_squared_error(y_test, pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, pred)

    print(f'alpha: {alpha:<4}  MSE:{mse:.3f}  RMSE:{rmse:.3f}  R2:{r2:.3f}')
    series = pd.Series(data=lasso.coef_, index=boston.feature_names)
    coef_df[f'alpha: {alpha}'] = series 

display(coef_df)
````

### 6-03. 릿지(Ridge) 모델 - Polynomial

   - alpha = 1.0 : L2 norm 사용 (Euclidean)
> image.png
- https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html   

````
# alphas = (0, 0.005, 0.1, 1)로 하여 모델 적용

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge

def sin(X):
    return np.sin(1.5 * np.pi * X)
dgr = 18
m = 30
X = np.sort(np.random.rand(m))
y = sin(X) + np.random.randn(m) * 0.1


alphas = (0, 0.005, 0.1, 1)
plt.figure(figsize=(15,5))

for i, alpha  in enumerate(alphas):
    polynomial = PolynomialFeatures(degree=dgr)
    ridge = Ridge(alpha=alpha)
    pipe = Pipeline([('polynomial', polynomial), ('ridge', ridge)])
    pipe.fit(X.reshape(-1,1), y)
    
    X_test = np.linspace(0,1, 100)
    plt.subplot(1, len(alphas), i+1)
    plt.plot(X_test, sin(X_test), 'g-', label='True Y')
    plt.plot(X_test, pipe.predict(X_test[:, np.newaxis]), 'r--', label='Predict')
    plt.plot(X, y, 'b.', label='Sample')
    plt.xlim((0,1)); plt.ylim((-2,2))
    plt.legend()
    plt.title(f"Degree:{dgr}, Alpha:{alpha}")
    
plt.show()
````
