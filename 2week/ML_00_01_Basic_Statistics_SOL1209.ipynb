## 1.통계학이란?

### 1-01. 모집단, 표본

  image.png

- 모집단(Population) : 잘 정의된 연구목적과 이와 연계된 명확한 연구대상 (데이터 전체 집합)
  - 예) 대통형 후보의 지지율 - 유권자
- 전수조사 : 모집단 전체를 대상으로 조사하는 것, 모집단은 매우 커 전체를 조사하기 어려움
- 표본(Sample) : 모집단으로부터 선택된 일부의 개체
  - 예) 각종 여론조사에 참여한 유권자
- 빅데이터에서는 모집단과 표본이 비슷하거나 같은 개념이 될 수 있음
  - 빅데이터는 전수조사!
- 모수(parameter) : 모집단의 특성을 나타내는 수치, 모집단의 평균, 분산 등
- 통계량(statistic) : 표본의 특성을 나타내는 수치, 표본의 평균, 분산 등

### 1-02. 통계학(Statistics)란?
- 관심 있는 연구의 대상인 **모집단의 특성을 파악**하기 위해
- 모집단부터 일부의 **자료(표본)를 수집**하고
- 수집된 표본을 정리, 요약, 분석하여 **표본의 특성을 파악**한 후
- **표본의 특성을 이용하여 모집단의 특성을 추론(inference)하는 원리와 방법을 제공하는 학문**

### 1-03. 표본(sample) 추출 방법
- 확률 표본 추출법(probability sampling) : 모집단을 구성하는 모든 추출단위에 대해 표본으로 추출된 활률을 알 수 있는 추출법
  - 표본 추출틀(sampling frame) 필요
  - 예) 모집단: {①, ②, ③, ④, ⑤} ⇒ 2개의 표본 ⇒ 어떤 개체가 표본으로 뽑힐 확률 2/5 
  - 단순확률추출, 계통추출, 집락추출, 층화추출 등
- 비확률 표본 추출법(non-probability sampling) : 특정 표본이 선정될 확률을 알 수 없음
  - 편의추출, 유의추출(전문가), 할당추출
  - 통계학에서 이야기하는 표본으로 사용할 수 없음

### 1-04. 확률 표본 추출법
- 단순무작위추출(simple random sampling)
   - 모집단의 각 개체가 표본으로 선택될 확률이 동일하게 추출되는 경우
   - 모집단의 개체수 N, 표본 수 n 일때 개별 개체가 선택될 확률은 n/N임
- 계통 추출(Systematic sampling)
   - 모집단 개체에 1, 2, ..., N 이라는 일련번호 부여 후, 첫 번째 표본을 임의로 선택하고 일정 간격으로 다음 표본을 선택함
   - 1 ~ 100 번호 부여후, 10개 선택한다면, [1, 11, 21, 31, ... , 91]  선택
- 층화추출(Stratified random sampling)
  - 모집단을 서로 겹치지 않게 몇 개의 집단 또는 층(strata)으로 나누고, 각 집단 내에서 원하는 크기의 표본을 단순 무작위추출법으로 추출함
  - 층 : 성별, 나이대, 지역 등
- 군집 추출(Cluster sampling)
  - 모집단을 특성에 따라 여러 개의 집단(cluster)로 나눔
  - 이들 집단 중 몇 개를 선택 한 후, 선택 된 집단 내에서 필요한 만큼의 표본을 임의로 선택함

### 2-03. 척도(Scale)의 종류
- 명목척도 : 단순히 측정 대상의 특성을 분류하거나 확인하기 위한 목적
  - 성별, 혈액형, 출생지
- 순서척도 : 대소 또는 높고 낮음 등의 순위(서열)만 제공할 뿐 양적인 비교는 할 수 없음
  - 금은동메달, 선호도, 만족도
- 등간척도(구간척도) : Interval Scale, 순위를 부여하되 순위 사이의 간격이 동일하여 양적인 비교 가능
   - 절대 0점이 존재하지 않음, 온도계수치, 물가지수
- 비율척도 : Ratio Scale, 절대 0점이 존재하여 측정값 사이의 비율 계산이 가능한 척도
   - 몸무게, 나이, 형제의 수, 직장까지 거리
- 절대 0점 : 없음을 의미함, 온도의 0도는 없음이 아니라 영상, 영하의 중간지점을 나타냄

```python
import numpy as np
import pandas as pd

bins = np.linspace(10.0, 100.0, 10)
print(bins, sep='\n')
labels = [f'{int(x)}~{int(x)+10}' for x in bins[:-1] ] 
cats = pd.cut(data_n, bins, right=False, labels=labels)
print(cats)
print('code:', cats.codes)
print('categories:', cats.categories)
```

```python
s = pd.Series(cats.codes).value_counts(sort=False)
s.index = cats.categories
s
```

```python
import numpy as np
import pandas as pd

bins = [10, 40, 50, 60, 70, 80, 100]

labels = []
intervals = []
for x in range(len(bins)-1):
    labels.append(f'{bins[x]}~{bins[x+1]}')
    intervals.append(bins[x+1]-bins[x])
print(bins, intervals, sep='\n')
cats = pd.cut(data_n, bins, right=False, labels=labels)
print(cats)
print('code:', cats.codes)
print('categories:', cats.categories)
```

```python
s = pd.Series(cats.codes).value_counts(sort=False)
s.index = cats.categories
s
```

```python
df2 = pd.DataFrame(s, columns=['도수'])
df2['계급구간길이'] = intervals
df2['상대도수'] = df2['도수']/df2['도수'].sum()
df2['누적상대도수'] = df2['상대도수'].cumsum()
df2
```

#### 3-03-3. 대체 통계값
- 표본 중앙값(median)

### 3-04. 산포(dispersion, 퍼짐)
- 자료들이 얼마나 퍼져 있는지를 나타내는 측도
- 중심위치가 얼마나 안정적인지에 대한 중요한 정보를 제공
  - 자료가 조밀하게 모임 : 중심위치의 변동성이 작아짐

```python
어떤 확률 변수의 분산(variance) : 확률변수가 기댓값으로부터 얼마나 떨어진 곳에 분포하는지 가늠하는 숫자
기댓값 : 확률변수의 위치
분산 : 기댓값이 얼마나 넓게 퍼져 있는지를 나타냄
표준 편차 : 분산의 제곱근으로 분포 분석에 더 많이 사용됨
```

```python
import scipy.stats as stats
import pandas as pd
import urllib
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
centers = [5, 5.3, 4.5]
colors = 'brg'

# 표준편차를 0.1로 지정
std = 0.1
data_1 = []
for i in range(3):
    # states.norm(평균, 표준편차).rvs(개수)
    data_1.append(stats.norm(centers[i], std).rvs(100))
    plt.plot(np.arange(len(data_1[i]))+i*len(data_1[0]),data_1[i], '.', color = colors[i])
```

```python
# 표준편차를 2로 지정
std_2 = 2
data_2 = []
for i in range(3):
    data_2.append(stats.norm(centers[i], std_2).rvs(100))
    plt.plot(np.arange(len(data_1[i]))+i*len(data_2[0]), data_2[i], '.', color = colors[i])
```

#### 3-04-8. 왜도 & 첨도
- 자료 분포의 형태를 나타내는 측도
- 모집단의 중심위치를 기준으로 대칭(symmetric)이라고 
가정한다
- 자료가 모집단 가정을 만족하는지 확인하는데 왜도 및 첨도를 사용한다
- 왜도(skewness) : 자료가 대칭적으로 분포되어 있는지 한쪽으로 기울어져 있는지에 대한 측도
   - 정규분포 : 0 skewness 
   - 오른쪽 치우침(긴꼬리), positive skewness : 왜도 > 0
   - 왼쪽 치우침(긴꼬리), negative skewness : 왜도 < 0



image.png

- 첨도(Krutosis) : 양쪽꼬리가 얼마나 두터운지를 나타내는 값
   - 정규분포 : 첨도 0 (Pearson 첨도 = 3)
   - 위로뾰족 : 첨도 > 0 (Pearson 첨도 > 3)
   - 아래로 퍼짐 : 첨도 < 0 (Pearson 첨도 < 3)

image.png
