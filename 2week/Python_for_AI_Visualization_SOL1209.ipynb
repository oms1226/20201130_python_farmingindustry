### 2-01. 선 그래프(Line plot)
- plot 설명에서 색, marker 등에 대한 정보 확인
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

````python
# 선 그래프 그리기에는 plt.plot() 함수를 사용한다

plt.figure(0, figsize=(6, 4))
x = np.arange(10)
plt.plot(x)
````

````python
# plt.plot(x, y) 함수를 사용하여 다음과 같은 그래프를 그려본다
# x, y 축을 갖는 그래프
# x 축 : 0, 1,2,3,...9
# y 축 : x ** 2
x = np.arange(10)
y = x ** 2
plt.plot(x,y)
````

````python
# 스타일 지정
# plt.plot(x, y, style) 함수를 사용하여 다음과 같은 그래프를 그려본다
# x, y 축을 갖고, 빨간색 원('ro')으로 표시되는 그래프
# x 축 : 0, 1,2,3,...9
# y 축 : x ** 2
# 스타일 : 빨간색 원
x = np.arange(10)
y = x ** 2
plt.plot(x,y, 'ro')
````

````python
# 여러 선 그리기
# 하나의 plot 함수에 넣기
# plt.plot(x, y, style, x2, y2, style2, ...)
# 'ro--', 'gs--', 'b^--' 사용

x = np.arange(10)
y = x**2
plt.plot(x, y, 'ro-', x, 2*y, 'gs--', x, 3*y, 'b^--')
````

````python
# 여러 선 그리기
# 위의 그래프를 plt.plot(x, y, style)을 여러번 사용하여 겹쳐 그리기로 작성해 보도록 한다
# 'ro--', 'gs--', 'b^--' 사용

x = np.arange(10)
y = x**2
plt.plot(x, y, 'ro-')
plt.plot(x, y*2, 'gs-')
plt.plot(x, y*3, 'b^--')
````

````python
# figure를 여러 개 지정하면 한 셀(cell)에 여러 개 그릴 수 있음
plt.figure(1, figsize=(7,5))
x = np.arange(10)
y = x**2
plt.plot(x, y, 'ro-')

plt.figure(2, figsize=(7,5))
plt.plot(x)
if 0 :
    plt.show()

if 1 :
    # 이미 선택했던 번호를 다시 선택하면 해당 그림에 추가 그림을 누적할 수 있음
    # 단 이미 show를 호출했으면 새 그림으로 그려짐
    plt.figure(1) 
    plt.plot(x, 2*y, 'gs--')
    plt.plot(x, 3*y, 'b^--')
    plt.show()
````

````python
# plt.xlabel(str) : x 축에 대한 레이블 표시
# plt.ylabel(str) : y 축에 대한 레이블 표시
# plt.title(str) : plot에 대한 제목

# 다음 출력과 같이 레이블과 제목을 표시하도록 한다.
x = np.arange(10)
y = x**2
plt.plot(x, 3*y, 'bs--')
plt.xlabel('time')
plt.ylabel('money')
plt.title('time for money')
````

````python
# Limitation : 전체 데이터 중에 특정 영역만 표시
# plt.xlim(min, max)
# plt.ylim(min, max)

x = np.arange(10)  # 0~9
y = x ** 2         # 0~81
plt.plot(x, y*3, 'bs--')   # 0~9, 0~243

# x 축에 대해서 0~5, y축에 대해서 0~80 영역 표시
plt.xlim(0, 5)
plt.ylim(0, 80)
````

````python
# plt.axis[xmin, xmax, ymin, ymax] 를 사용해 표시 영역 설정
x = np.arange(10)  # 0~9
y = x ** 2         # 0~81
plt.plot(x, y*3, 'bs--')   # 0~9, 0~243
# x 축에 대해서 0~5, y축에 대해서 0~80 영역 표시
plt.axis([0,5, 0, 80])
````

````python
# Legend(범례), 범례 설정을 위해 그래프에 label이 있어야 함
# plt.plot(x, y, style, label='text')
# plt.legend(loc='text')
# 또는
# plt.plot에 label을 생략하고, plt.legend(labels=[label1, label2 ...]) 를 사용한다 (순서, 개수 동일)
# loc : upper left, upper center, upper right, center left, center, center right
# lower left, lower right, best, right 등이 있음

x = np.arange(10)
y = x ** 2

# 다음과 같은 결과로 범례가 표시되도록 plt.plot의 label, plt.legend의 loc을 지정해 보도록 한다
# y는 y, 2*y, 3*y 를 사용하고, 스타일은 'ro--', 'gs--', 'b^--'를 사용한다
if 0 :
    plt.plot(x, y, 'ro--', label='red')
    plt.plot(x, 2*y, 'gs--', label='green')
    plt.plot(x, 3*y, 'b^--', label='blue')
    plt.legend(loc='best')
else :    
    plt.plot(x, y, 'ro--')
    plt.plot(x, 2*y, 'gs--')
    plt.plot(x, 3*y, 'b^--')
    plt.legend(loc='best', labels=['red', 'green', 'yellow'])
````

````python
# 그림 내 텍스트
# Annotation
# plt.annotate(text, xy, xytext, arrowprops) # 매우 다양한 형식이 있음!
# annotations : https://matplotlib.org/tutorials/text/annotations.html
# - text : 표시할 문구
# - xy : 화살표를 표시할 좌표
# - xytext : 문구를 표시할 좌표
# - arrowprops : 화살표 속성

# plt.text(x, y, text)
# - x, y : 문구 표시할 위치 좌표
# - text : 표시할 문구
x = np.arange(10)
y = x ** 2
plt.plot(x, y, 'ro--')
plt.plot(x, 2*y, 'gs--')
plt.plot(x, 3*y, 'b^--')
plt.annotate('here', xy=(7,150), xytext=(6, 175), arrowprops={'color':'orange'})
plt.text(6.5, 225, '$y=3x^2$')
plt.text(7.7, 160, '$y=2x^2$')
plt.text(7.9, 80, '$y=x^2$')
plt.grid()
````

### 2-02. Subplot
- 여러 plot을 나누어 표시
- plt.subplot(row, col, idx)
   - row, col : shape(행, 열)
   - idx : 일련번호 1부터 시작
- figure, axes = plt.subplots(row, col, **args) 
   - row, col : shape(행, 열)
   - sharex = True (x축을 공유한다는 의미) 
   - constrained_layout = True (subplot 간 간격 확보 - 제목 달때)
- figure : 1개 캔버스
- axes : 캔버스 위에 그려지는 1개 그래프
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html  

image.png

````python
# subplot
x = np.arange(10)
y = x ** 2
plt.subplot(2,2,1)
plt.plot(x, y, 'ro--')
plt.axis([0, 8, 0, 75])
plt.grid()
plt.subplot(2,2,2)
plt.plot(x, 2*y, 'gs--')
plt.axis([0, 8, 0, 75])
plt.grid()
plt.subplot(2,2,3)
plt.plot(x, 3*y, 'b^--')
plt.axis([0, 8, 0, 75])
plt.grid()
plt.subplot(2,2,4)
plt.plot(x, 4*y, 'm-')
plt.axis([0, 8, 0, 75])
plt.grid()
````

````python
# subplot - iteration 사용
x = np.arange(10)
y = x ** 2
r, c = 2, 2

styles = ['ro--', 'gs--', 'b^--', 'm-']
# 위와 동일한 동작을 하는 subplot을 iteration을 사용하여 만들어 보도록 합니다.
for n, style in zip(range(1, r*c+1), styles) :
    plt.subplot(r, c, n)
    plt.plot(x, y*n, style)
    #plt.xticks([]) # xticks 표시 하지 않음
    #plt.yticks([])  # yticks 표시 하지 않음
    #plt.axis('off') # axis를 아예표시 하지 않음 - 이미지 같은 것 그릴 때 사용
    plt.axis([0, 8, 0, 75])
    plt.grid()

plt.show()
````

````python
# subplots
x = np.arange(10)
y = x ** 2
r, c = 2, 2

fig, axes = plt.subplots(r, c)#, sharex=True, sharey=True)   
styles = ['ro--', 'gs--', 'b^--', 'm-']
for axflat, n, style in zip(axes.flat, range(1,r*c+1), styles[:r*c]):
    #axflat.set(xticks=[], yticks=[])  # 축 ticks 없앨때 사용
    #axflat.axis('off') # axis를 아예표시 하지 않음
    axflat.plot(x, y*n, style)
    axflat.axis([0, 8, 0, 75])
    axflat.grid()
plt.show()
````

````python
# subplot - 일부만 표시
# subplot (행, 열, 번호)
x = np.arange(10)
y = x ** 2
plt.subplot(4,4,1)
plt.plot(x, y, 'ro--')
plt.axis([0, 8, 0, 75])
plt.grid()
plt.subplot(4,4,3)
plt.plot(x, 2*y, 'gs--')
plt.axis([0, 8, 0, 75])
plt.grid()
plt.subplot(4,4,12)
plt.plot(x, 3*y, 'b^--')
plt.axis([0, 8, 0, 75])
plt.grid()
````

### 2-03. Scatter(산점도, 산포도, 분산 형 차트)
- 두 데이터간의 관계를 보기 위해 사용함
- plt.scatter(x, y, s, c, alpha)
   - s : scale
   - c : color
   - alpha : 투명도(0~1)
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html   
- line style은 링크 참고
  - https://matplotlib.org/gallery/lines_bars_and_markers/linestyles.html#sphx-glr-gallery-lines-bars-and-markers-linestyles-py   

````python
x = np.arange(50)
y = x + np.random.rand(50) * 10
m = np.random.randint(1, 100, 50)
c = np.random.randint(0, 2, 50)

# 크기를 m, 색상을 c, 투명도를 0.7로 지정하여 산점도 그래프를 그리세요. 
plt.scatter(x, y, s = m, c = c, alpha = 0.7)
````

````python
# scatter (산포도) : 데이터 분포를 보기 위해 사용하는 그래프
x = np.arange(50)
y = x + np.random.randn(50) * 10
m = np.random.randint(1, 100, 50)
c = np.random.randint(0, 2, 50)

fig = plt.figure()
# s=size, c=color, alpha=투명도 (0~1, 0:투명, 1:불투명)
plt.scatter(x, y, s=m, c=c, alpha=0.7)

# figure().add_subplot(1,1,1) -> axes 반환
ax = fig.add_subplot(1,1,1)

xmajor_ticks = np.arange(0, 51, 10)
xminor_ticks = np.arange(0, 51, 2)
ymajor_ticks = np.arange(-30, 71, 10)
yminor_ticks = np.arange(-30, 71, 5)

ax.set_xticks(xmajor_ticks)
ax.set_xticks(xminor_ticks, minor=True)
ax.set_yticks(ymajor_ticks)
ax.set_yticks(yminor_ticks, minor=True)


plt.grid(which='both', color='gray', dashes=(3,3))
````

````python
# scatter (산포도)
# 데이터 분포를 보기 위해 사용하는 그래프
x = np.arange(50)
y = x + np.random.randn(50) * 10
m = np.random.randint(1, 100, 50)
c = np.random.randint(0, 2, 50)

# s=size, c=color, alpha=투명도 (0~1, 0:투명, 1:불투명)
fig, ax = plt.subplots(1,1)

plt.scatter(x, y, s=m, c=c, alpha=0.7)
xmajor_ticks = np.arange(0, 51, 10)
xminor_ticks = np.arange(0, 51, 5)   #  여기를 5로 변경
ymajor_ticks = np.arange(-30, 71, 10)
yminor_ticks = np.arange(-30, 71, 5)
ax.set_xticks(xmajor_ticks)
ax.set_xticks(xminor_ticks, minor=True)
ax.set_yticks(ymajor_ticks)
ax.set_yticks(yminor_ticks, minor=True)
plt.grid(which='both', color='gray', dashes=(3,3))
````

### 2-05. 히스토그램(Histogram)
- 히스토그램을 그리기 위한 hist 명령은 bins 인수로 데이터를 집계할 구간 정보를 받음
- 데이터 집계 결과를 반환
- n, bins, patches = plt.hist(x, bins, range)
   - x : 데이터
   - bins : 계급
   - range(대상 범위)
   - n : 도수 
   - patches : 계급별 히스토그램 좌표와 높이
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html   

````python
# histogram
x = np.random.normal(100, 15, 10000)
# 반환 받는 것은 필수 아님
n, bins, patches = plt.hist(x, 50)
plt.grid()
plt.show()
````

````python
print(n, bins, patches, sep='\n')
````

````python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.text(60, 0.025, f'$\mu={mu},\ \sigma={sigma}$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
````

### 3-01. Line Plot

````python
# Seaborn으로 라인플롯(Line plot) 그리기
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2 * np.sin(x+0.5)
y3 = 3 * np.sin(x+1.0)
y4 = 4 * np.sin(x+1.5)

plt.figure(figsize=(10,6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
````

#### 3-02-1. BoxPlot으로 분포 확인

````python
# 값의 분포를 boxplot으로 확인함
# x에 1D 값을 전달하면 수평, y에 1-D 값을 전달하면 수직

sns.set_style("whitegrid")

plt.figure(figsize=(8,6))
sns.boxplot(x=tips["total_bill"])
#sns.boxplot(y=tips["total_bill"])
plt.show()
````

#### 3-03-4. heatmap
- 두 개의 카테고리 값에 의한 값의 변화를 보고자 할 경우에 사용
- 표출하고자 하는 데이터를 먼저 데이터 테이블 형태로 만들어야 함

````python
# flights 데이터 세트 불러오기
flights = sns.load_dataset("flights")
flights.head(5)
````

````python
# 해당 데이터를 heatmap으로 만들기 위해 'pivot' 함수를 사용하여 데이터 테이블 생성
flights = flights.pivot("month", "year", "passengers")
flights
````

````python
# heapmap 함수를 사용하여 데이터 테이블 시각화
plt.figure(figsize=(10,5))
sns.heatmap(flights)
plt.show()
````

````python
# annot 옵션을 통해 값 출력 가능
# fmt 옵션을 통해 데이터의 형식을 지정(d:정수, f:실수)
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
````

````python
iris.corr()
````

````python
# pairplot을 이용하여 변수(컬럼)간 관계와 분포 확인 가능
sns.set(style="ticks")
sns.pairplot(iris)
plt.show()
````

````python
# hue 옵션을 이용하여 카테고리별 분포를 볼 수  있음
sns.pairplot(iris, hue='species')
plt.show()
````

## 5.다양한 Plot 예

### 5-01. 표준 정규분포(normal distribution) 

image.png

````python
import numpy as np

x = np.linspace(-5, 5, 101)   # x 정의
# 정규분포 : N(μ, σ/σ2)으로 표시
# 표준정규분포의 평균 μ는 0, 분산/표준편차 σ/σ2는 1 이다
y = (1 / np.sqrt(2 * np.pi)) * np.exp(- x ** 2 / 2 )
#display(x, y)
````

````python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
plt.plot(x, y)                       
plt.xlabel("x")                      # x축 레이블 지정
plt.ylabel("y")                      # y축 레이블 지정
plt.grid()                           # 플롯에 격자 보이기
plt.title("Normal Distribution without scipy")     # 타이틀 표시
plt.legend(["N(0, 1)"])              # 범례 표시
plt.show()                           # 플롯 보이기
````

````python
import scipy.stats as stats

# stats.norm(평균, 표준편차).pdf(x) 를 사용하여 y 작성
y1 = stats.norm(0, 1).pdf(x)        
plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정
plt.plot(x, y1, color="red")         # 선을 빨강색으로 지정하여 plot 작성          
plt.xlabel("x")                      # x축 레이블 지정
plt.ylabel("y")                      # y축 레이블 지정
plt.grid()                           # 플롯에 격자 보이기
plt.title("Normal Distribution with scipy.stats")     # 타이틀 표시
plt.legend(["N(0, 1)"])              # 범례 표시
plt.show()                           # 플롯 보이기
````

````python
import scipy.stats as stats

legend = []
y1 = stats.norm(0, 1).pdf(x)     
def norm (mu, std) :
    legend.append("N(" + str(mu) + ", " + str(std) + ")")
    return stats.norm(mu, std).pdf(x)        

plt.figure(figsize=(10, 6))          # 플롯 사이즈 지정

# 표준편차를 1 ~ 4로 지정함
for i in np.arange(1, 5):
    plt.plot(x, norm(0, i))          # plot 추가       
    
plt.xlabel("x")                      # x축 레이블 지정
plt.ylabel("y")                      # y축 레이블 지정
plt.grid()                           # 플롯에 격자 보이기
plt.title("Normal Distribution with scipy.stats")     # 타이틀 표시
plt.legend(legend)                   # 범례 표시
plt.show()                           # 플롯 보이기
````
