### 1-01. 함수의 필요성, 용어, 정의  
- 특정 작업을 수행하는 명령문의 그룹
- 프로그램의 기능 단위로 모듈을 구성하는데 사용됨
- 프로그램이 복잡해지고 커질 때 함수를 사용하면 체계적이고 관리도 쉬워짐
- 코드의 반복을 피할 수 있고, 코드의 재사용이 가능함
    
- 참고 : https://youtu.be/EoUCOihCFG0 (깨봉수학 함수의 정의 참조)
- 용어
    - argument : 인수, 함수에 전달하는 객체, 실체(instance, object)
    - parameter : 매개변수, 인수를 가리킬 이름 (인수를 저장할 변수)
    - return : 반환 
- 함수의 정의
    - ```def 함수명(paragmeter_list) :```
        - ```''' doc string '''```
        - ```statements```
        - ```return 반환값```    


- 함수의 구성요소
    - 키워드 def는 함수의 시작을 표시함
    - 함수명은 식별자(identifier) 작성 규칙을 따름
    - parameter의 사용은 선택 사항임
    - 함수 헤더의 끝은 콜론(:)으로 표시함
    - 문서 문자열(docstring)은 함수가 수행하는 작업을 설명하는 것으로 선택 사항임
    - 함수 본문은 하나 이상의 statement로 작성됨
    - **statement는 동일한 들여쓰기가 적용되어야 함**
    - 함수에서 값을 반환하는 return문은 선택 사항임
- 함수 호출
    - 함수를 정의한 후에는 다른 함수, 프로그램 등에서 호출
    - 함수를 호출하기 위해서는 argument를 전달해야 함
- return
    - return 문은 함수를 종료하고 함수를 호출한 곳으로 돌아가는 데 사용
    - return 문은 expression을 반환 함
    - return 문 뒤에 expresion이 없거나, return 문 자체가 없으면 함수는 None 반환

### 1-03. argument의 종류
- 함수의 호출시 사용되는 argument    
    - positional argument : 위치에 따라 값이 전달되는 방식
    - keyword argument : = 를 사용하여 특정 parameter에 값을 제공함, 순서에 상관 없음
- 함수 정의시 사용되는 argument
    - default argument : 새로운 argument가 전달되지 않을 경우 사용되는 기본 값
- Python의 argument rule
     1. 함수 정의 시 : non-default argument가 앞, default argument가 뒤에 위치해야 한다.
     2. 함수 호출 시 : positional argument가 앞, keyword argument가 뒤에 위치해야 한다.

````
def function(a, b):  # a, b : parameter
    return [a, b]

# 어떻게 사용하는 것일까요?
x = function()   # positional argument
y = function()   # keyword argument
print(x, y)
````

````
# default argument  (b = 0)
def mySum(a, b = 0):   # def mySum(a = 1, b):  (불가능)
  return a + b

mySum(10), mySum(10, 5)
````

````
# Error - positional argument follows keyword argument
def func(a, b, c):
  return a + b + c

func(a=10, b=20, c)    
````

````
# Error - non-default argument follows default argument
def func(a=1, b=2, c) : 
  return a + b + c

func(10)
````

### 1-04. parameter의 종류
어떤 종류의 argument를 전달 할 수 있는가?!
- positional or keyword parameter   ```(a, b, c)```
    - parameter 앞쪽으로 *, 뒤쪽으로 / 가 없음
- positional only parameter  ```(a, b, /)```  # v3.8 부터
    - parameter 뒤쪽으로 /가 있음
- keyword only parameter   ```(*, a, b)  (*a, b, c)```
    - parameter 앞쪽으로 *가 있음 (이름과 떨어진 상태로)
- var-positional parameter  ```(*a)```
    - parameter 이름에 *가 하나 붙어 있음
    - tuple 형태로 전달된다
- var-keyword parameter  ```(**a)```
    - parameter 이름에 *이 두개 붙어 있음
    - dict 형태로 전달된다

> var는 0개 이상의 가변적 argument를 의미한다

````
# positional or keyword parameter
def func(a) :
    print(a)

func(10)    # positional argument
func(a=20)  # keyword argument
````

````
# positional only parameter
# python v3.8 이상에서 사용 가능
def func(a, b, /) :
    print(a)
  
func(10, 20) # positional argument
````

````
# keyword only parameter
def func(*, a, b) :
    print(a)

func(b=20, a=10)    # keyword argument 
````

````
# Quiz - 각 parameter의 종류가 무엇인지 적어보세요
# a : 
# b, c : 
def func(a, *, b, c) :
  return a + b + c    # ( a = 10, b = 20, c = 30)

#func(10, b = 20, c = 30)
#func(a=10, b = 20, c = 30)
func(c = 30, b = 20, a=10)
````

````
# var positional parameter
def func (*a):
    print(type(a), a)

func(1,2,3,4,5,6,7)
func()    
````

````
# a : var positional parameter, b : keyword only parameter
def func (*a, b):
    print(type(a), a)
    print(b)

func(1,2,3,4,5,6,7, b = 10)
func(b=20)  
````

````
# var-keyword parameter : 맨 마지막 parameter로만 사용가능
def func(**a) :
    print(type(a), a)

func(A=10, B=20, C=30, D=40)    
func()
````

### 1-07. namespace 종류
- built-in namespace
  - 1개 존재하며 built-in 함수, exception 목록 등이 포함되어 있음
  - print, input, zip, map, enumerate, 
  - int, float, str, tuple, list, set, dict ...

- global namespace
  - 모듈(=파일) 마다 1개 존재
  - 해당하는 모듈(파일) 전체에서 사용할 수 있는 이름 관리
  - globals() => dict 반환

- local namespace
  - 함수(메서드) 마다 1개 존재
  - 해당 함수(메서드)에서만 사용할 수 있는 이름 관리
  - locals() => dict 반환

- namespace의 특징
  - namespace는 이름(name)을 관리하는 dict 자료구조이다
  - dict는 mutable 자료구조이기 때문에 삽입, 갱신, 삭제 등이 가능하다
  - 추가, 갱신 : assignment
  > A = 100
- name resolution : namespace에서 이름을 찾는 규칙
  - 현 위치에서 가장 가까운 namespace부터 상위로 올라가면 찾는다
  - 함수 내부 : local -> global -> built-in
  - 함수 외부 : global -> built-in
- 이름(name)의 제거는 이름 앞에 del 키워드를 사용해서 한다     

image.png

````
A = 100
C = 500
def func() :
  A = 200
  B = 300
  print(f'local  : {A}') # 200
  print(f'global : {C}') # 500  - local namespace에 C가 없기때문에 global namespace에서 C를 찾아 사용  
  print(f'local namespace : {locals()}')

print = 1000  # global namespace에 'print'이름이 1000이라는 정수 객체와 쌍으로 등록
del print     # namespace에서 print라는 이름을 제거
print(f'global : {A}') # 100

print(func.__code__.co_varnames) 
func()
````

### 6-03. for문의 사용
    for target_list in iterator :
        simple/compound statement
    else :
        simple/compound statement

````
for statement -> 반복을 위해 iterator가 필요하다
1) in 뒤에 있는 것에 대해 iter() 함수를 사용하여 iterator를 얻는다
   => in 뒤에는 iterable, iterator를 사용할 수 있다  (iterator는 __iter__가 있어서 iterable이라고 할 수 있음)
2) next() 를 실행해 처리할 대상을 가져온다
3) 대상에 target(이름)을 부여한다, target(이름)은 for 와 in 사이에 나열한다
4) target을 사용하여 처리한다
5) 반복동작은 next()가 StopIteration을 발생하면 멈춘다 (더이상 처리할 데이터가 없으면)
6) break에 의한 종료가 아니면 else에 포함된 문장을 실행한다
````

````
colors = ['red', 'green', 'blue']  # colors is an iterator (X)
x = iter(colors)  # x is an iterator
print(next(x))
````

````
colors = ['red', 'green', 'blue']  # colors is an iterator (O, X)
for c in colors:
    print(c)

print(c)  # 마지막에 c에 저장되었던 값이 기억되어 있음
````

````
# 홀수를 출력하다 짝수가 나오면 반복을 종료한다
a = (1, 2, 3)
for x in a:
     if not x % 2 : break
     print(x)
else:  # 첫 번째 suite 에서 break에 의해 종료 되지 않으면 실행됨 
     print("done1")
````

````
# os 모듈의 getcwd 함수 import
from os import getcwd
print(getcwd())
````

````
# os 모듈의 getcwd 함수를 pwd 라는 이름으로 import
from os import getcwd as pwd
print(pwd())
````

````
# from 모듈이름 import *은 모듈 내부의 모든 함수를 import 하는 것임
from os import *
print(getcwd())
````

### 8-04. numpy 라이브러리 import 예

````
# numpy 라이브러리를 import, 이름을 numpy로 사용

print(numpy.nan)
````

````
# numpy 라이브러리를 import, 이름을 np로 사용 (numpy라는 이름은 사용하지 않음)

print(np.nan)
````

````
# numpy 라이브러리 내의 random 모듈만 import, 이름을 random으로 사용
# numpy에 포함된 다른 속성, 함수 사용 불가

print(random.randint(1, 10))
````

````
# numpy 모듈내의 random 모듈만 import, 이름을 rd로 사용

print(rd.randint(1, 10))
````

````
# numpy 라이브러리 내의 모든 이름을 import 
  
print(random.randint(1, 10), nan)
````

### 8-05. 모듈의 작성

````
%%writefile my_calc.py
def add(a, b) :
    return a+b

def sub(a, b) :
    return a-b
 
````

````
# import 모듈명

a, b = 20, 10
print(my_calc.add(a, b), my_calc.sub(a, b) )
````

````
# import 모듈명 as 모듈닉네임

a, b = 20, 10
print(mc.add(a, b), mc.sub(a, b))
````

````
# from 모듈명 import 함수명1, 함수명2 ...

a, b = 20, 10
print(add(a, b), sub(a, b))
````

````
# from 모듈명 import *

a, b = 20, 10
print(add(a, b), sub(a, b))
````
