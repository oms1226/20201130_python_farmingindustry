### 1-02. Variable, Sample
- sample : 사건(event), 데이터 분석에 사용되는 하나의 데이터 단위, 1개 sample은 여러 개의 variable로 구성될 수 있음
- variable : feature라고도 많이 사용하며, 어떤 sample이 가진 속성(attribute)을 의미함, 너무 많은 variable은 분석에 방해가 됨
- data수는 (sample, variable) 형태의 shape으로 표기함
- sample > variable 관계이여야 함
- 머신러닝에서 입력에 사용되는 variable을 독립변수, 출력에 사용되는 variable을 종속변수라고 함
- 종속변수가 '연속형'인 경우 회귀분석, '범주형'인 경우 분류 학습을 수행할 수 있음

### 1-04. 데이터 세트(Data Set)
- Data Set : 머신러닝에서 입력(X), 출력(y)에 사용되는 데이터 묶음
- Data Set는 1개 이상의 입력과 1개의 출력으로 구성됨
- 지도(supervised) 학습에서 입력으로 사용되는 것이 data, 출력으로 사용되는 것이 target
  - data : **2차원 구조**이며, 1개 sample에 대한 data는 1차원 구조 임
  - target : **1차원 구조**, 1개 sample에 대한 target은 1개 scalar 값
- feature_names : 각 feature에 대한 이름 (column 명)
- target_names : target의 종류에 대한 이름 목록
- data와 target은 numeric type 데이터이어야 함

  - Hold-out: 성능 검증을 위해 Data Set을 Training Set, Test Set으로 분리하여 사용함

   - 기계가 알아서 학습하여 결과를 찾아내는 방법
   - 입력은 주어지지만 출력은 제공되지 않음

### 2-02. Overfitting
- 머신러닝 진행 시 주의해야되는 가장 중요한 문제 중 하나
- Overfitting 은 너무 잘 맞아 떨어진다는 의미임
- Training data에만 너무 잘 맞아 Training data 이외의 데이터들에 대해서는 잘 맞아 떨어지지 않는 경우
- 해결 방법
  - Cross validation
  - Regularization
  - Remove Features
  - Early stopping
  - Ensembling
  - penalty hyperparameters 조절
- underfitting : 아직 학습이 제대로 이루어지지 않은 상태
- 적절한(appropriate) 학습이 좋음!

### 2-03. 파라미터(Parameter)와 하이퍼파라미터(Hyperparameter)
- 머신러닝 모델의 Parameter :  모델의 구성요소, 데이터로부터 학습되는 것
- 머신러닝 모델의 Hyperparameter : 모델 학습 과정에 반영되며, 학습을 시작하기 전에 미리 값을 결정하는 것
- Parameter : Linear Regression에서 y = **W**x + **b** 와 같은 직선 방정식의 W, b를 찾는 것
- Hyperparameter : kNN에서 k의 개수, Ridge, Lasso의 α값, Learning Rate 등

### 5-04. 유방암 LogisticRegression

````
'malignant', 'benign' # 악성, 양성 판단
````

````
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# 데이터 가져오기
cancer = load_breast_cancer()

# 데이터 전처리 - Scaling
scaler = StandardScaler()
scaled = scaler.fit_transform(cancer.data)

# 데이터 전처리 - 데이터 셋 분리(Training, Test data 분리)
X_train, X_test, y_train, y_test = train_test_split(scaled, cancer.target, random_state=0)

# 모델 학습 (Training data 사용)
lr_clf = LogisticRegression()
lr_clf.fit(X_train, y_train)

# 모델 예측 (Test data 사용)^^ 안
pred = lr_clf.predict(X_test)

# 모델 평가 (정확도(accuracy) 평가)
print(f'train accuracy : {lr_clf.score(X_train, y_train)*100:.2f}%')
print(f'test accuracy  : {accuracy_score(y_test, pred)*100:.2f}%')
````
