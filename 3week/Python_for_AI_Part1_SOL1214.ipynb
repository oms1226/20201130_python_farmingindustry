### 1-04. class, instance, object는 무엇일까?!
- 파이썬 프로그램에서는 주로 'instance'라고 불리는 것으로 명령을 수행한다
- instance는 class (=type)에 의해서 생성되는 object를 의미한다
- class는 type이라고 하는 class에 의해 생성되는 object를 의미한다
- class는 instance를 abstraction 해서 컴퓨터에 저장한 object이다
    - class는 instance가 사용할 공통 attribute/method를 제공
    - instance는 attribute/method를 사용할 때 필요한 자신만의 value를 가짐 

- object : class(type), instance 를 모두 통칭    

````
# 다음 작업을 해볼까요!

import math

number = int(input())
answer = math.sqrt(number)
print(answer)
````

### 1-05. object의 특징
- object는 메모리에 저장된 것으로 자신만의 주소를 갖는다
- object 마다 자신의 속성(attritute, value) 및 기능(method)을 가지고 있다
   - 속성 및 기능은 '.' 연산을 이용하여 사용할 수 있다
- 다음의 함수를 사용하여 정보를 확인할 수 있다   
   - dir(object) : 자주 사용되는 속성 및 기능 목록 확인
   - id(object) : 저장된 주소 확인
   - type(object) : type 확인   
- 동일한 이름의 속성, 기능이라도 object 별로 다르게 동작할 수 있다
- special method는 __이름__ 와 같이 앞/뒤에 __(던더, double underscore)가 붙어 있다
   - special method는 '.' 연산을 사용하기 보다 연산자 및 built-in 함수로 대신하여 사용한다

````
# 10 값을 갖는 정수 객체에 대해 id, type, dir을 확인한다
a = 10
display(id(a), type(a))
print(dir(a))
````

````
# 같은 연산이라도 정수와 문자열은 다른 동작을 한다
# + 에 대한 special method 는 __add__ 이다
a, b = 10, 20
print(a + b, a.__add__(b))
a, b = '10', '20'
print(a + b, a.__add__(b))
````

### 3-01. 산술 연산자
- ** : 거듭제곱
  - pow(A, B) : A ** B
  - pow(2, 4) => 16
- +, -, *, / : 더하기, 빼기, 곱하기, 나누기
- //, % : 몫, 나머지
  - divmod(A, B) : A를 B로 나눈 몫, 나머지를 tuple로 반환하는 함수
  - divmod(10, 3) => (3, 1)
- 우선 순위 : 거듭제곱 > 곱하기, 나누기, 몫, 나머지 > 더하기 빼기



````
# 사칙연산 구해보기
a, b = 10, 3
print(a+b, a-b, a*b, a/b)
````

````
#연산 우선 순위에 주의해야 하며, 우선 순위 변경을 위해 소괄호 ( ) 를 사용함

a = 3 + 4 * 2
b = (3 + 4) * 2
print(a, b)
````

````
# 연산자는 사용하기 간편하지만, 정확하게 사용해야하며, 이름/함수를 사용하면 실수를 줄이거나 유연한 사용 가능

print((-1)**2, -1**2)
a, b = -1, 2
print(a**b, pow(a, b))
````

````
# a의 값을 1 증가시켜 출력해 보자
a = 10
a = a + 1   # a += 1
print(a)
````

````
# a의 값을 b만큼 증가시켜 출력해 보자
a, b = 10, 20
a += b
print(a)
````

````
# 연산자를 사용하여 a를 b로 나눈 몫, 나머지를 구하여 출력해 보자
a, b = 10, 3
print(a//b, a%b, divmod(a, b))
````

````
# divmod 함수를 사용하여 a를 b로 나눈 몫, 나머지를 구하여 출력해 보자
# name unpack을 사용하여 몫은 m, 나머지는 n의 이름을 갖도록 한다 
a, b = 10, 3
m, n = divmod(a, b)  # name unpack
print(m, n)
````

````
# container 활용
a, b = 10, 3
x = divmod(a, b)
print(*x)          # argument unpack
print(x[0], x[1])  # subscription
````

````
# [Q] 다음 c의 각 자리를 분리하여 출력해 보자
c = 123
print(c//100, c//10%10, c%10)
````

````
# Numeric type 끼리는 type이 달라도 연산 가능
# 표현 범위가 넓은 type의 결과를 생성함  float > int > bool
# int + float = float, int + bool = int
a = 10 + 3.14
b = 1 + True
print(a, type(a))
print(b, type(b))
````

### 3-02. 비교 연산자
- 결과가 True/False 이다
- ```==, !=``` : **값이** 같다, 다르다
- is, is not : **객체가** 같다, 다르다 (id() : 객체의 주소, 고유값)
   - is : id(a) == id(b),  is not : id(a) != id(b)
- ```>, <, >=, <=``` : 크다(초과), 작다(미만), 크거나 같다(이상), 작거나 같다(이하)
- membership 연산 
  - in, not in : 값이 container에 포함되어 있다, 없다

````
# a 와 b가 같은지 비교하여 결과를 출력하자
a, b = 5, 10
print(a==b, a!=b)
````

````
# a와 b가 같은 객체인지 비교하여 결과를 출력하자
# interning : 같은 값을 갖는 경우 객체 재사용
a = 10
b = 10
print(a is b, a is not b)
````

### 4-10. 문자열 포맷팅
- 형식(format)을 사용하여 문자열을 만드는 방법
- % 연산자, f-string, str.format() 사용
    - https://docs.python.org/3/library/string.html#format-string-syntax


````
x = 100

# 'x is 100'
print('x is', 100)

# 포맷 지시자를 포함한 문자열 % (객체)
print('x is %d' % (x))  
# f-string
print(f'x is {x}')
# 문자열의 format 메서드
print('x is {}'.format(x))
````

````
five = 5
# %, f-string, str.format을 사용하여 다음을 출력하세요.
# 'I eat 5 apples.'
print('I eat %d apples.' % (five))
print(f'I eat {five} apples.')
print('I eat {} apples.'.format(five))
````

## 6.Container - Mappings type

### 6-01. dict의 특징 및 생성
- 딕셔너리는 key와 value의 쌍 여러 개와 { }로 구성됨
    - dishes = {'eggs':2, 'sausage' :1, 'bacon':1, 'spam':500}
- key를 사용하여 value를 얻음
- item에 순서가 없음
- key는 고유 값을 설정해야하며 hashable이어야 함    
  - hashable : immutable한 객체 (str, tuple, int, float, bool )
- 다양한 생성자 : 함수에 대한 parameter 학습 필요
  - dict()
  - dict(**kwargs)
  - dict(mapping, **kwargs)
  - dict(iterable, **kwargs)    

````
# key는 hashable만 사용할 수 있으며
# hashable에는 int, float, bool, tuple, str 등이 있다.
# 다양한 type의 key를 갖는 dict 객체를 생성해 보자!
a = {1: 'one', 'two': 2, (1, 2) : 'tuple', 3.14 : 'pi'}
a[3.14]
````

````
# 다음 dict 객체의 출력을 예측해보자
d = { 'one' : 1, 'one' : 2, 'one': 3}

d
````

````
# numeric의 True, 1, 1.0 은 같은 hash 값을 갖는다
# numeric의 False, 0, 0.0 은 같은 hash 값을 갖는다

# 다음의 결과를 예측하세요! (^_^!!)
x = {True : "one", 1 : "One", 1.0 : "ONE"}
x
````

### 6-02. dict의 연산
- key를 사용하여 value 얻기 : r = d[key]
- 추가, 갱신 : d[key] = value
   - key가 존재하면 갱신, 존재하지 않으면 추가
- 삭제 : del d[key]
- 얻기, 삭제 동작시 존재하지 않는 key를 사용하면 KeyError 발생

````
# dict 생성
a = {'age' : 28, 'height' : 180}
a
````

````
# a에서 'age'에 해당하는 value 얻기
a['age']
````

````
# dict 객체에 새로운 item 추가 방법/ 갱신 방법
# dict[key] = object 를 사용한다

# a 에 'name'을 key로 하고 'tom'을 value하는 item을 추가한다
a['name'] = 'tom'
a
````

````
# 'age'를 30으로 갱신하여 보도록 한다

a['age'] = 30    
a
````

````
# del 삭제할_value  을 사용하여 'name' 을 key로 하는 item을 삭제하여 본다 
# 'name':'tom' item을 삭제 한다

del a['name']
a
````

````
a = {'name':'tom', 'phone':'010-1234-5678', 'birth': '2001-1-2'}
a
````

````
a['name']  # KeyError
````

````
# dict.get(key, default=None) : key에 해당하는 item의 value를 반환한다 (삭제는 하지 않음)
# key에 해당하는 item이 없으면 default를 반환한다. default 지정하지 않으면 None 반환
# a 에서 'name'의 value를 반환 받아 출력한다
x = a.get('name', -1) # code here
if x == -1 :
  print('there is no "name"')
````

````
# 'height'를 key로 하는 item의 value를 반환받아 출력, default는 -1로 지정한다
x = a.get('height', -1)# code here
x
````

````
# dict.update(dict object) dict.update(key=value) : dict 객체의 value를 갱신하거나, item을 추가한다
a = {'name':'KIM'}
b = {'gender' : 'F', 'age' : 25}
# a에 포함된 name을 'yoon'으로 갱신한다
a.update(name='yoon')
# a 뒤에 b 추가한다 (gender, age 는 없는 key이므로 추가)
a.update({'gender' : 'F', 'age' : 25})
print(a)
````

````
# dict.fromkeys(iterable[,v])
# iterable의 item을 key로 하고, value를 v로 하여 새사전 생성하여 반환, v를 주지 않으면 None을 사용함
# 새로운 사전을 생성하는 생성자 역활을 하는 메서드

d = dict.fromkeys(list("0123456789"), 0)
print(d)
````

````
# dict.setdefault(key, defaultValue) : key 값이 존재하지 않을 때만 값을 설정한다
e = {"A" : 90, "B" : 80, "F" : 50}
e.setdefault("D", 60)
e.setdefault("B", 85)
print(e)
````

## 8.입출력 함수

- ```print(*value, sep=' ', end='\n', ...)```
    - *value : 0개 이상, 콤마로 나열하여 전달
    - sep : value들 사이 분리 기호
    - end : 출력 문자열 마지막 문자
- ```a = input(prompt=None)```
    - prompt : 입력 안내를 위한 메시지 (필수 아님) 
    - 반환값은 str 이므로 필요에 따라 형변환하여 사용   

````
print()
print(10)
print(10, 20) # =>  10 20
print(10, 20, sep=', ') # =>  10, 20
print(10, 20, sep='-')  # =>  10-20 
print(10, end='')
````
