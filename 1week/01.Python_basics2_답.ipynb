```python
#반복문 조금 더 살펴보기
nums = [0, 1, 2, 3, 4]
```

```python
squares = [ ]   #빈 리스트
print(squares) 
```

```python
# 반복문 확장 문법1
nums = [0, 1, 2, 3, 4]
squares = [x ** 2 for x in nums]  # 파이썬스러운 문법!
print(squares) 
```

```python
# 반복문 확장 문법2 
# 반복문 + 조건문 : 짝수인 값의 제곱을 구해서 even_squares 리스트에 넣기
even_squares = []
nums = [0, 1, 2, 3, 4]
for x in  nums :
    if x % 2 == 0 :
        even_squares.append(x**2)

print(even_squares)
```

```python
nums = [0, 1, 2, 3, 4]
even_squares = [ x**2    for x in  nums if x % 2 == 0 ]
print(even_squares)
```

```python
# 반복문 확장 문법2 
# 반복문 + 조건문 : 짝수인 값의 제곱을 구해서 even_squares 리스트에 넣기
nums = [0, 1, 2, 3, 4]
even_squares = [x ** 2 for x in nums if x % 2 == 0] 
print(even_squares) 
```

### 실습 : Python_basics ④

```python
score = [75, 46, 93, -82, -77, 130, 100, 75, -10]

# 1) score에서 음수를 제거한 리스트를 score01 변수에 담으세요. 
score01 = [ ]   #빈 리스트
for i in  score  :
    if i >= 0 :
        score01.append(i)
         
print(score01)
```

```python
# 2) score01에서 100보다 큰 수를 제외한 리스트를 score02 변수에 담으세요.
score02 = []

for i in score01 :
    if i <= 100 :
        score02.append(i)

print(score02)
```

```python
# 3) score02 리스트의 element들 합과 평균을 구하시오

print(sum(score02), sum(score02) / len(score02))

```

```python
# 4) score02 리스트 각 element의 값을 10으로 나눈 값을 갖는 리스트 score03을 만드세요.

score03 = [ ]   #빈 리스트
for i in score02 :
    score03.append(i /10)

print(score03)
```

### 09.함수(Function)


```python
# 함수는 input, 처리, output

def sign(x):
    if x > 0:
        return 'positive'
    elif x < 0:
        return 'negative'
    else:
        return 'zero'

for x in [-1, 0, 1]:
    print(sign(x))
    
```

```python
def hello(name, loud=False):  #디폴트 파라미터는 순서상 뒤쪽
    if loud:
        print('HELLO, {}!'.format( name.upper() )  )
    else:
        print('Hello, {}'.format(name) )

hello('Bob')              # "Hello, Bob"
hello('Fred', loud=True)  # "HELLO, FRED!"
```

```python
hello('hanky', loud = False)
```

### 실습 : Python_basics ⑥

```python
# 1) 숫자의 list를 받아서 평균을 반환하는 average 함수를 만들어 보자. 
def average(list_score) :
    # 여기를 작성해 주면 됩니다. 
    # 힌트: list_points의 요소의 평균을 계산해서 avg 변수에 담으면 됩니다. 
    avg = sum(list_score) / len(list_score)
    return avg

score = [50, 60, 70, 80, 90, 100]
a = average(score)
print(a)
```

```python
# 2) 1)번에서 만든 함수를 수정하시오 : 만약 빈(항목이 없는) 리스트가 입력될 경우 
# "리스트에 값이 없습니다."를 반환하도록 수정
def average(list_score):
    # 여기를 작성해 주면 됩니다. 
    # 힌트: list_points의 요소의 평균을 계산해서 avg 변수에 담으면 됩니다. 
    if list_score == [] :
        return "리스트에 값이 없습니다."
    else :
        avg = sum(list_score) / len(list_score)
        return avg

blank_list = [1,2,3,4]
a = average(blank_list)
print(a)
```

```python
# 3) 1 이상의 자연수 n과 m을 입력받아, 
#     n보다 작은 자연수 중 m의 배수를 
#    리스트에 저장하고 
#    리스트를 출력하는 함수를 작성하시오.
#    옵션, 리스트에 저장하는 부분은 확장문법으로 작성하시오.
def fn01(n, m) :
    list_a = []
    for i in range(1,n) :
        if i % m ==0 :
            list_a.append(i)
    return list_a
    
```
