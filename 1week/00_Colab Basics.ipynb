## 4. 코드 Line Number

아래 셀을 선택하고 Ctrl+M L를 눌러보세요


```python
# 실행해봅시다.
sample_list = []
sumsum = 0

for i in range(10) :
  sample_list.append(i)
  sumsum += i
  print(i)
print('loop done')
print(sumsum)
print(sample_list)
```
