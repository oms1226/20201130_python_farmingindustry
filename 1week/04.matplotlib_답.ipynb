### Axis Label, Title

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

plt.show()
```

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

# Ticks
plt.yticks([0, 2, 4, 6, 8, 10]
           , ['0', '2B', '4B', '6B', '8B', '10B'])

plt.show()
```

### Common Axis

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]
pop3 = [3.231, 2.3, 1.988, 0.334]

plt.plot(year, pop1, 'red')
plt.plot(year, pop2, 'blue')
plt.plot(year, pop3, 'green')

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

plt.show()

```

### Subplot

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]

# plot1
plt.subplot(2, 1, 1)
plt.plot(year, pop1, 'red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop1')

# plot2
plt.subplot(2, 1, 2)
plt.plot(year, pop2, 'blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop2')

plt.tight_layout() # 두 그래프 사이 여백 주기
plt.show()
```

## 4.matplotlib with pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
aq = pd.read_csv("https://raw.githubusercontent.com/DA4BAM/dataset/master/airquality_simple.csv")

# 날짜를 yyyymmdd로 만든 후 변환해봅시다.
aq['Date'] = pd.to_datetime(1973*10000 + aq.Month*100 + aq.Day
                              ,format='%Y%m%d')

aq.head()
```

* Line chart

```python
aq.plot("Date", "Ozone", color = 'red')
plt.show()
```

* Scatter

```python
aq.head()
```

```python
aq.plot(kind = "scatter", x= "Temp", y= "Ozone")
plt.show()
```

```python
# wind와 Ozone
aq.plot(kind = "scatter", x= "Wind", y= "Ozone")
plt.show()
```

```python
# Solar.R과 Ozone
aq.plot(kind = "scatter", x= "Solar.R", y= "Ozone")
plt.show()
```

* histogram

```python
aq.plot(kind = "hist", y= "Ozone", bins = 20, grid =1)
```

* Barplot with group by

```python
aq.groupby("Month")['Ozone','Temp'].mean().plot(kind = 'bar')
plt.show()
```

* boxplot

```python
aq.plot(kind = 'box', y = "Wind")
```

```python
aq.boxplot("Wind", by = 'Month')
```

## 실습 

```python
import pandas as pd
```

```python
# 데이터 준비
g_price = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/garak_market_price.CSV')
g_price['Date'] = pd.to_datetime(g_price['Date'], format='%Y%m%d')
g_price['Month'] = g_price['Date'].dt.month
g_price['UnitPrice'] = g_price['Price'] / g_price['KG']
```

```python
g_price.head()
```

```python
# 1) 라인차트를 그려봅시다.
# 고구마, 특급 상품을 조회하여 df1에 저장하시오.
# 고구마, 상급 상품을 조회하여 df2에 저장하시오.

df1 = g_price[(g_price['Grade']==0) & (g_price['ProductName'] == '고구마') ]
df2 = g_price[(g_price['Grade']==1) & (g_price['ProductName'] == '고구마') ]

# df1에 대해서 일별 고구마 단가 추이를 라인차트로 그려봅시다.

df1.plot('Date', 'Price', figsize=(15,8))
plt.show()
```

```python
# 2) 고구마 특급, 상급 상품의 단가 추이를 라인차트로 그려봅시다.

plt.figure(figsize=(15, 8))
plt.plot(df1['Date'], df1['Price'])
plt.plot(df2['Date'], df2['Price'])
plt.show()
```

```python
# 3) 고구가 특급 상품의 단가에 대해서 월별로 박스플롯을 그려봅시다.

df1.boxplot("Price", by = 'Month', figsize=(15,8))
plt.show()
```

```python
# 4) 고구마 상급 상품의 단가에 대해서 히스토그램을 그려봅시다.

df2.plot(kind = 'hist', y = "Price", bins = 30, grid = 1, figsize = (15,8))
plt.show()
```
