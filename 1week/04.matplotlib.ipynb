## 2.연속형 데이터의 분포

### Histogram

```python
values = [0,1.6,1.4,2.2,2.5,2.6,3.2,3.5,3,3.9,4.2,6]
plt.hist(values, bins = 8)#bins : 구간의 갯수
plt.show()
```

### Boxplot

```python
values = [0,1.6,1.4,2.2,2.5,2.6,3.2,3.5,3,3.9,4.2,6]
plt.boxplot(values)#동그라미는 찐 max 값이다.
plt.show()
```

```python
#횡 방향으로 박스플롯 그리기
plt.boxplot(values, 0, None, 0)  #네번째 파라미터가 방향을 나타냄. 
plt.show()
```

### Density Plot

```python
import seaborn as sns

sns.distplot(values, hist=True, kde=True, bins=6)#표준정규분포가 density plot 중 하나이다!

#sns.distplot(values, hist=True, kde=True, 
#             bins=6, color = 'darkblue', 
#             hist_kws={'edgecolor':'black'},
#             kde_kws={'linewidth': 4})
```

### Axis Label, Title

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

plt.show()
```

### Ticks

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]

plt.plot(year, pop)

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

# Ticks
plt.yticks([0, 2, 4, 6, 8, 10]
           , ['0', '2B', '4B', '6B', '8B', '10B'])

plt.show()
```

### Common Axis

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]

plt.plot(year, pop1, 'red')
plt.plot(year, pop2, 'blue')

# label
plt.xlabel('Year')
plt.ylabel('Population')

# Title
plt.title('World Population Projections')

plt.show()

```

### Subplot

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]

# plot1
plt.subplot(2, 1, 1)
plt.plot(year, pop1, 'red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop1')

# plot2
plt.subplot(2, 1, 2)
plt.plot(year, pop2, 'blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop2')

plt.show()
```

```python
year = [1950, 1970, 1990, 2010]
pop1 = [2.519, 3.692, 5.263, 6.972]
pop2 = [1.231, 2.252, 2.988, 4.334]

# plot1
plt.subplot(2, 1, 1)
plt.plot(year, pop1, 'red')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop1')

# plot2
plt.subplot(2, 1, 2)
plt.plot(year, pop2, 'blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Pop2')

plt.tight_layout() # Or equivalently,  "plt.tight_layout()" #두 그래프에 간격주기!

plt.show()
```

## 실습 

```python
import pandas as pd
```

```python
# 데이터 준비
g_price = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/garak_market_price.CSV')
g_price['Date'] = pd.to_datetime(g_price['Date'], format='%Y%m%d')
g_price['Month'] = g_price['Date'].dt.month
g_price['UnitPrice'] = g_price['Price'] / g_price['KG']
```

```python
g_price.head()
```

```python
# 1) 라인차트를 그려봅시다.
# 고구마, 특급 상품을 조회하여 df1에 저장하시오.
df1 = g_price.loc[(g_price['ProductName'] == '고구마') & (g_price['Grade'] == 0 )]

# 고구마, 상급 상품을 조회하여 df2에 저장하시오.
df2 = g_price.loc[(g_price['ProductName'] == '고구마') & (g_price['Grade'] == 1 )]

# 고구마이면서 특급(0)의 월별 평균 가격을 조회해 봅시다.
df1.groupby('Month', as_index = False)['Price'].mean()



```

```python
df1.head()
```

```python
df1.plot("Date", "UnitPrice", figsize=(15,8))
```

```python
# df1에 대해서 일별 고구마 단가 추이를 라인차트로 그려봅시다.

df1.plot("Date", "UnitPrice", color = 'red')
plt.show()
```

```python
# 2) 고구마 특급, 상급 상품의 단가 추이를 라인차트로 그려봅시다.
df1.plot("Date", "UnitPrice", color = 'red')
df2.plot("Date", "UnitPrice", color = 'blue')
plt.show()
```

```python
# 2) 고구마 특급, 상급 상품의 단가 추이를 라인차트로 그려봅시다.
plt.figure(figsize=(15,8))
plt.plot(df1["Date"], df1["UnitPrice"], color = 'red')
plt.plot(df2["Date"], df2["UnitPrice"], color = 'blue')
plt.show()
```

```python
# 3) 고구가 특급 상품의 단가에 대해서 월별로 박스플롯을 그려봅시다.
df1.boxplot("UnitPrice", by = 'Month')


```

```python
# 4) 고구마 상급 상품의 단가에 대해서 히스토그램을 그려봅시다.
df2.plot(kind = "hist", y= "UnitPrice", bins = 80, grid =True, figsize = (15,8))


```
