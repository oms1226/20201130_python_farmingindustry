### 2.7 Join(Merge)


```python
# 두 데이터셋을 불러옵니다.
g_in = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/garak_market_in.CSV')
g_price = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/garak_market_price.CSV')
```

```python
# 복습1 : g_price에서 칼럼을 추가하시오
# 1kg 당 가격 (Unit_Price)
g_price['Unit_price'] = g_price.Price / g_price.KG
g_price.head()
```

```python
# 복습2 : 위 결과에 이어서, g_price에서 Date 별, ProductID별 평균 Unit_Price를 구하여 g_price2에 저장하시오.
g_price2 = g_price.groupby(['Date', 'ProductID'], as_index=0)['Unit_price'].mean()
g_price2.head()
```

```python
# 복습3 : 칼렴 명 Unit_Price를 Avg_Price로 변경하시오.
g_price2.rename(columns={'Unit_price':'Avg_Price' }, inplace=1)
g_price2.head()
```

* 판다스에서 join은 굉장히 간단합니다.
* 자동으로 key를 잡아줍니다
* default로 inner join을 합니다.

```python
g_in.head()
```

```python
g_price2.head()
```

```python
# 공통된 칼럼이 있으면 그 칼럼을 key로 사용
# 위 head() 결과에서 공통된 칼럼은 무엇인가요?

# Date, ProductID
```

```python
# 이제 merge를 시도해 봅시다.

g_in_price = pd.merge(g_in, g_price2 )
g_in_price.head()
```

### 실습 : Join(Merge)

조금 복잡한 데이터 가지고 연습을 해 봅시다.  
Merge는 아래와 같은 상황에서 주로 사용됩니다!

* sales : 판매 정보
* products : 상품정보


```python
import pandas as pd
```

```python
sales = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/sales.csv')
sales.head()
```

```python
products = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/products.csv')
products.head()
```

```python
# sales를 가지고 집계를 먼저 해 봅시다.
# 상품ID별로 매출액(Amt)의 합계를 구해 봅시다.
sales.groupby('ProductID', as_index=0)['Amt'].sum()
```

```python
# 2) 상품이름(ProductName)별로 매출액(Amt)의 합계를 구해봅시다.
# 그리고 매출액이 높은 순으로 정렬해 봅시다.
sales2 = sales.groupby('ProductID', as_index=0)['Amt'].sum()
pd.merge( products.loc[:, ['ProductID', 'ProductName']], sales2 )
```

```python
pd.merge( products, sales2 ).sort_values('Amt', ascending = 0)
```

```python
# 3) Category 별로 매출액을 집계하고 매출액이 높은 순으로 정렬해 봅시다.
sales3 = pd.merge(sales, products)

sales4 = sales3.groupby('Category', as_index = 0)['Amt'].sum()

sales4.sort_values('Amt', ascending = 0)
```

```python
# 3) Category 별로 매출액을 집계하고 매출액이 높은 순으로 정렬해 봅시다.
pd.merge(sales, products).groupby('Category', as_index = 0)['Amt'].sum().sort_values('Amt', ascending = 0)
```

### 2.9 Concat

데이터 프레임을 위/아래로, 양 옆으로 붙여 봅시다.


```python
# 고구마이면서 특급(0)의 월별 평균 가격을 조회해 봅시다.
g_price.loc[(g_price['ProductName'] == '고구마') & (g_price['Grade'] == 0 )].groupby('Month', as_index = False)['Price'].mean()
```

```python
g_price_go_0 = g_price.loc[(g_price['ProductName'] == '고구마') & (g_price['Grade'] == 0 )]

g_price_go_0.groupby('Month', as_index = False)['Price'].mean()
```

```python
# 고구마이면서 grade 0, 1 등급의 월별 평균 가격을 각각 저장해 봅시다.
sp_0 = g_price.loc[(g_price['ProductName'] == '고구마') & (g_price['Grade'] == 0 )].groupby('Month', as_index = False)['Price'].mean()
sp_1 = g_price.loc[(g_price['ProductName'] == '고구마') & (g_price['Grade'] == 1 )].groupby('Month', as_index = False)['Price'].mean()
```

```python
sp_0
```

```python
sp_1
```

```python
# 칼럼 이름 수정
sp_0.columns.values[1] = '특_평균가격'
sp_0.head()
```

```python
sp_1.columns.values[1] = '상_평균가격'
sp_1.head()
```

```python
# 이제 concat 열 기준으로 붙여 봅시다.
pd.concat([sp_0 , sp_1], axis=1)
```

```python
# 월이 중복되는군요. 오른쪽은 제외하고 합쳐봅시다.
pd.concat([sp_0 , sp_1.iloc[:,1]], axis=1)
```
